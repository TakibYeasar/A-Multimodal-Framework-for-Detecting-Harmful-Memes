```python
train_data = pd.read_json("./data/raw/HM Dataset/train.jsonl", lines=True)
dev_data   = pd.read_json("./data/raw/HM Dataset/dev.jsonl", lines=True)
test_data  = pd.read_json("./data/raw/HM Dataset/test.jsonl", lines=True)
```

### Fix Image Paths

```python
data_dir = "./data/raw/HM Dataset/"

def fix_paths(df):
    df['img_path'] = df['img'].apply(lambda x: os.path.join(data_dir, x))
    return df
```

```python
train_data = fix_paths(train_data)
dev_data   = fix_paths(dev_data)
test_data  = fix_paths(test_data)
```

### Save as CSV for next notebook

```python
train_data.to_csv("./data/processed/train_metadata.csv", index=False)
dev_data.to_csv("./data/processed/dev_metadata.csv", index=False)
test_data.to_csv("./data/processed/test_metadata.csv", index=False)
```
